```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics
import seaborn as sns
```

```python
df = pd.read_csv("size_new.csv")
```

```python
df.head()
```

```python
def histogram_with_std(df, column):
  x = df[column]
  plt.hist(x, bins=20)
  plt.gca().set(title=column, ylabel='Count'); 
  m = statistics.mean(x)
  sd = statistics.stdev(x)
  plt.axvline(m, color='k', linestyle= 'dashed')
  plt.axvline(m+sd , color='y', linestyle= 'dashed')
  plt.axvline(m-sd , color='y', linestyle= 'dashed')
  plt.axvline(m+(2*sd) , color='g', linestyle= 'dashed')
  plt.axvline(m-(2*sd), color='g', linestyle= 'dashed')
  plt.axvline(m + (3*sd), color='r', linestyle= 'dashed')
  plt.axvline(m - (3*sd), color='r', linestyle= 'dashed')
```

```python
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
#Red line is 3 std away from mean. The data after red line is outlier
# Plot Histogram on x
histogram_with_std(df, 'count')
```

```python
def draw_box_plot(df, col):
  data = df[col]
  fig = plt.figure(figsize =(8, 6))
  # Creating plot
  plt.boxplot(data)
  # show plot
  plt.show()
```

```python
draw_box_plot(df, "count")
```
